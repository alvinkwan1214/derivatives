```python
from derivatives import eu_option, futures, normal_stock
import numpy as np
import matplotlib.pyplot as plt 
```

# Cover Cell

```python
underlying_range = np.linspace(20, 60, 41)
payout_call = eu_option(50, underlying_range, option_type="Call", position='short', cost=5)

payout_stock = normal_stock(40, underlying_range)

total = payout_call + payout_stock
plt.plot(underlying_range, total, 'b', label='Overall')
plt.plot(underlying_range, payout_stock , 'g', linestyle='dashed', label="stock")
plt.plot(underlying_range, payout_call , 'r', linestyle='dashed', label="Selling Call")
plt.legend()
plt.axhline(0, color='black', linewidth=1)
plt.grid()
```

# Protective Put

```python
underlying_range = np.linspace(20, 60, 41)
payout_put = eu_option(43, underlying_range, option_type="Put", position='long', cost=5)

payout_stock = normal_stock(40, underlying_range)

total = payout_put + payout_stock
plt.plot(underlying_range, total, 'b', label='Overall')
plt.plot(underlying_range, payout_stock , 'g', linestyle='dashed', label="stock")
plt.plot(underlying_range, payout_put , 'r', linestyle='dashed', label="Buying Put")
plt.legend()
plt.axhline(0, color='black', linewidth=1)
plt.grid()
```

# Straddle 

```python
underlying_range = np.linspace(20, 60, 41)
strike_price = 40
cost_call = 5
cost_put = 6

payout_call = eu_option(strike_price, underlying_range, option_type="Call", position='long', cost=cost_call)
payout_put = eu_option(strike_price, underlying_range, option_type="Put", position='long', cost=cost_put)

total_payout_straddle = payout_call + payout_put

plt.plot(underlying_range, total_payout_straddle, 'b', label='Straddle Payout')
plt.plot(underlying_range, payout_call, 'g', linestyle='dashed', label="Call Option")
plt.plot(underlying_range, payout_put, 'r', linestyle='dashed', label="Put Option")
plt.axhline(0, color='black', linewidth=1)
plt.grid()
plt.legend()
plt.xlabel('Underlying Stock Price')
plt.ylabel('Payout')
plt.title('Straddle Spread Payout Profile')
plt.show()
```

# Strangle

```python
import numpy as np
import matplotlib.pyplot as plt

underlying_range = np.linspace(20, 60, 41)
strike_call = 45
strike_put = 35
cost_call = 4
cost_put = 3

payout_call = eu_option(strike_call, underlying_range, option_type="Call", position='long', cost=cost_call)
payout_put = eu_option(strike_put, underlying_range, option_type="Put", position='long', cost=cost_put)

total_payout_strangle = payout_call + payout_put

plt.plot(underlying_range, total_payout_strangle, 'b', label='Strangle Payout')
plt.plot(underlying_range, payout_call, 'g', linestyle='dashed', label="Call Option")
plt.plot(underlying_range, payout_put, 'r', linestyle='dashed', label="Put Option")
plt.axhline(0, color='black', linewidth=1)
plt.grid()
plt.legend()
plt.xlabel('Underlying Stock Price')
plt.ylabel('Payout')
plt.title('Strangle Spread Payout Profile')
plt.show()
```

# Butterfly

```python
import numpy as np
import matplotlib.pyplot as plt

underlying_range = np.linspace(20, 60, 41)
strike_low = 35
strike_middle = 40
strike_high = 45
cost_low = 8.2
cost_middle = 8
cost_high = 9

payout_call_low = eu_option(strike_low, underlying_range, option_type="Call", position='long', cost=cost_low)
payout_call_middle = eu_option(strike_middle, underlying_range, option_type="Call", position='short', cost=cost_middle)
payout_call_high = eu_option(strike_high, underlying_range, option_type="Call", position='long', cost=cost_high)

total_payout_butterfly = payout_call_low + 2 * payout_call_middle + payout_call_high

plt.plot(underlying_range, total_payout_butterfly, 'b', label='Butterfly Spread Payout')
plt.plot(underlying_range, payout_call_low, 'g', linestyle='dashed', label="Long Call (Low Strike)")
plt.plot(underlying_range, payout_call_middle, 'r', linestyle='dashed', label="Short 2x Call (Middle Strike)")
plt.plot(underlying_range, payout_call_high, 'm', linestyle='dashed', label="Long Call (High Strike)")
plt.axhline(0, color='black', linewidth=1)
plt.grid()
plt.legend()
plt.xlabel('Underlying Stock Price')
plt.ylabel('Payout')
plt.title('Butterfly Spread Payout Profile')
plt.show()
```

# Iron Condor

```python
import numpy as np
import matplotlib.pyplot as plt

underlying_range = np.linspace(20, 60, 41)
strike_put_long = 30
strike_put_short = 35
strike_call_short = 45
strike_call_long = 50
cost_put_long = 1
cost_put_short = 3
cost_call_short = 3.5
cost_call_long = 2

payout_put_long = eu_option(strike_put_long, underlying_range, option_type="Put", position='long', cost=cost_put_long)
payout_put_short = eu_option(strike_put_short, underlying_range, option_type="Put", position='short', cost=cost_put_short)
payout_call_short = eu_option(strike_call_short, underlying_range, option_type="Call", position='short', cost=cost_call_short)
payout_call_long = eu_option(strike_call_long, underlying_range, option_type="Call", position='long', cost=cost_call_long)

total_payout_iron_condor = payout_put_long + payout_put_short + payout_call_short + payout_call_long

plt.plot(underlying_range, total_payout_iron_condor, 'b', label='Iron Condor Payout')
plt.plot(underlying_range, payout_put_long, 'g', linestyle='dashed', label="Long Put")
plt.plot(underlying_range, payout_put_short, 'r', linestyle='dashed', label="Short Put")
plt.plot(underlying_range, payout_call_short, 'm', linestyle='dashed', label="Short Call")
plt.plot(underlying_range, payout_call_long, 'y', linestyle='dashed', label="Long Call")
plt.axhline(0, color='black', linewidth=1)
plt.grid()
plt.legend()
plt.xlabel('Underlying Stock Price')
plt.ylabel('Payout')
plt.title('Iron Condor Payout Profile')
plt.show()
```
